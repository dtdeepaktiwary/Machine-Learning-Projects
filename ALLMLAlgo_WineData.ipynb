# UNSUPERVISED ALGORITHMS

### Using KMEANS Algorithm

```python
X_scaled.shape
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
#Assuming if the no of clusters are 2

model=KMeans(n_clusters=2,init='k-means++',max_iter=300)
model.fit(X)
model.inertia_
```

```python
#Assuming if the no of clusters are 3


model=KMeans(n_clusters=3,init='k-means++',max_iter=300)
model.fit(X)
model.inertia_
```

```python
#Assuming if the no of clusters are 4


model=KMeans(n_clusters=4,init='k-means++',max_iter=300)
model.fit(X)
model.inertia_
```

```python
#So instead for calculating inertia one by one we will try to calculate in a certain range of values for k

loss=[]
for i in range (1,10):
    model= KMeans(n_clusters=i,init='k-means++',max_iter=100)
    model.fit(X)
    loss.append(model.inertia_)
    print('Loss of the model is:',loss)
```

```python
plt.plot(loss)
plt.xlabel('no of clusters')
plt.ylabel('loss')
plt.title("Elbow method for best value of K")
plt.show()
```

```python
# so let us consider as value of k=3 as best no of clusters


model=KMeans(n_clusters=3,init='k-means++',max_iter=300)
model.fit(X)
model.inertia_
```

```python
model.labels_
```

```python
cen=model.cluster_centers_
cen[:,0]
```

```python
X.head(1)
```

```python
# Plotting the clusters now with their centroids


fig,axes=plt.subplots(2,2,figsize=(15,7))

sns.scatterplot(x='alcohol',y='proline',data=X,hue=model.labels_,palette='Set1',ax=axes[0,0])
sns.scatterplot(x=cen[:,0],y=cen[:,12],marker='*',s=500,color='Black',ax=axes[0,0])

sns.scatterplot(x='ash',y='proline',data=X,hue=model.labels_,palette='Set1',ax=axes[0,1])
sns.scatterplot(x=cen[:,2],y=cen[:,12],marker='*',s=500,color='Black',ax=axes[0,1])

sns.scatterplot(x='magnesium',y='alcalinity_of_ash',data=X,hue=model.labels_,palette='Set1',ax=axes[1,0])
sns.scatterplot(x=cen[:,4],y=cen[:,3],marker='*',s=500,color='Black',ax=axes[1,0])

sns.scatterplot(x='diluted_wine',y='proline',data=X,hue=model.labels_,palette='Set1',ax=axes[1,1])
sns.scatterplot(x=cen[:,11],y=cen[:,12],marker='*',s=500,color='Black',ax=axes[1,1])
```
