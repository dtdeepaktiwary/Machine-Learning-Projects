```python
# we will convert all the words into tokens and assign the index values of that words in the dictionary

token= Tokenizer()

token.fit_on_texts([faqs])


# We can check what index value is assigned to each word.
print(token.word_index)

# Total no of unique words in our dictionary
print(len(token.word_index))
```

```python
# Now we will check every sentence in faqs will be tokenized and then we will create list of list as a input data and outpu data
input_sequence =[]


for sentence in faqs.split('\n'):
  tokenized_sentence=token.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequence.append(tokenized_sentence[:i+1])

```

```python
input_sequence
```

```python
# We will check what is the maxmimum length of the longest sentence

print(max([len(x) for x in input_sequence]))


#How many sentences  are there in the sequence
print(len([len(x) for x in input_sequence]))
```

```python
# We need to apply padding by using max_length=57 to every sentence

from keras.preprocessing.sequence import pad_sequences

padded_input_sequences=pad_sequences(input_sequence,padding='pre',maxlen=57)

padded_input_sequences
```

```python
# Now that our input data is ready to be split into X and Y just to make this as look like supervised model of input and output

X= padded_input_sequences[:,:-1]

X
```

```python
y= padded_input_sequences[:,-1]

y
```

```python
# Since we are using the multiclass classification algo of prediction by supervised we will one_hor encode the output as vectors of 282+1 classes

from keras.utils import to_categorical

y= to_categorical(y,num_classes=283)
```

```python
print(X.shape)

print(y.shape)
```

# Creating the RNN Architecture


```python
model = Sequential()

model.add(Embedding(283,100,input_length=56))

model.add(LSTM(150))

# model.add(Dense(units=256,activation='relu'))

# model.add(Dense(units=512,activation='relu'))
# keras.layers.Dropout(rate=0.1)

model.add(Dense(units=283,activation='softmax'))

model.summary()
```

```python
#Compiling the model.

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#Training the model

model.fit(X,y,epochs=100,batch_size=100)
```

# Since the model is trained now we need to Test and Check predictions

```python
# Checking predictions on the random sample text.

text='About the'

# 1. tokenize
tokenized_text=token.texts_to_sequences([text])[0]


# 2.padding
padded_tokenized_text= pad_sequences([tokenized_text],maxlen=56,padding='pre')


# 3. Then checking prediction
pos=np.argmax(model.predict(padded_tokenized_text))
pos
```

```python
# Checking the next word should be:

for index,word in token.index_word.items():
  if index==pos:
    print(word)
```

```python
# If we want to have the next 5 word predictions for any word.

text='Checkout'

for i in range(6):
# 1. tokenize
 tokenized_text=token.texts_to_sequences([text])[0]


# 2.padding
 padded_tokenized_text= pad_sequences([tokenized_text],maxlen=56,padding='pre')


# 3. Then checking prediction
 pos2=np.argmax(model.predict(padded_tokenized_text))

 for index,word in token.index_word.items():
   if index==pos2:
     text=text +' '+ word
     print(text)
```
