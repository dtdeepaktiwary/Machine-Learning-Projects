```python
# We will tokenize all the statements in the given faqs as our machine dosent understands words.

token = Tokenizer()

token.fit_on_texts([faqs])

#Checking the index values of each words
print(token.word_index)

print()
print(token.index_word)
print()

#checking the no of unique words in the vocab
print(len(token.word_index))
```

```python
# Generating the list of the statements so that it can be sequentially tokenized.

input_sequences =[]

for sentence in faqs.split("\n"):
  tokenized_sentence=token.texts_to_sequences([sentence])[0]

  for i in tokenized_sentence:
    input_sequences.append(tokenized_sentence[:i+1])

input_sequences
```

```python
# Now we will do the padding of each tokenized sentence and before that we need to check the sentence with maximum length

# The sentence with maximum length has 17 words in it.
print(max([len(x) for x in input_sequences]))

print()

#There are total of 3479 sentences
print(len([len(x) for x in input_sequences]))
```

```python
# Now we will do the padding

from keras.preprocessing.sequence import pad_sequences

padded_input_sequence = pad_sequences(input_sequences,padding='pre',maxlen=17)

print(padded_input_sequence)

print()

print(np.shape(padded_input_sequence))
```

```python
print(padded_input_sequence[0])
print()
print(padded_input_sequence[1])
print()
print(padded_input_sequence[2])
```

```python
# Splitting our data into X and y so that we can covert padded_input into supervised kind of data to train

X =padded_input_sequence[:,:-1]
y= padded_input_sequence[:,-1]

print(np.shape(X))
print(np.shape(y))
```

```python
print(X[0])

print()
print(y[0])
```

```python
# Now we need to one_hot encode our target variable into vector shape of 3479+1
from keras.utils import to_categorical

y= to_categorical(y,num_classes=3480)
```

# Creating the RNN Architechture

```python
model = Sequential()

model.add(Embedding(1134,100,input_length=16))

model.add(LSTM(1134))

model.add(Dense(units=256,activation='relu'))
keras.layers.BatchNormalization()
model.add(Dense(units=512,activation='relu'))
keras.layers.Dropout(rate=0.1)


model.add(Dense(units=3480,activation='softmax'))

model.summary()
```

```python
# Compiling the model

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=50,batch_size=200)
```
